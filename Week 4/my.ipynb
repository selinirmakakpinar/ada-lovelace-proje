# Solution:

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://realpython.github.io/fake-jobs/"
text = requests.get(url).content.decode('utf-8')
```

```python
soup = BeautifulSoup(text, 'html.parser')
```

```python
type(soup)
```

```python
soup.title
```

```python
print(soup.prettify())
```

```python
soup.find_all('a')
```

```python
count=1
for link in soup.find_all('h2'):
    print(f"{count} : {link}")
    count+=1
```

```python
array=[]
for x in soup.find_all('h2'):
    array.append(x)
```

```python
print(array)
```

```python
for i in array:
    if "Python" in str(i):
        print(str(i))
```
